```python
# PROBABILIDAD Y ESTADÍSTICA CON PYTHON
# Análisis de regresión lineal y correlación para grupo1, grupo2 y grupo3

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr

# -------------------------------------------------------------------
# Función para interpretar la fuerza de la correlación
# -------------------------------------------------------------------
def interpretar_fuerza_correlacion(r):
    r_abs = abs(r)
    if r_abs < 0.3:
        return "muy débil"
    elif r_abs < 0.6:
        return "moderada"
    elif r_abs < 0.8:
        return "fuerte"
    else:
        return "muy fuerte"

# -------------------------------------------------------------------
# Función principal de análisis por grupo
# -------------------------------------------------------------------
def analizar_grupo(nombre_grupo, archivo_csv):
    print("=" * 70)
    print(f"ANÁLISIS {nombre_grupo}")
    print("=" * 70)

    # 1) Carga de datos
    df = pd.read_csv(archivo_csv)

    # Renombrar columnas si fuera necesario (descomenta si tus nombres varían)
    # df.columns = ["HorasEstudio", "Calificacion"]

    print("\nPrimeras filas del dataset:")
    print(df.head())

    # 2) Exploración de datos
    print("\nDescripción estadística:")
    print(df.describe())

    # 3) Histograma de calificaciones
    plt.figure()
    plt.hist(df["Calificacion"], bins=10, edgecolor="black")
    plt.title(f"Histograma de Calificaciones - {nombre_grupo}")
    plt.xlabel("Calificación")
    plt.ylabel("Frecuencia")
    plt.tight_layout()
    plt.show()

    # 4) Modelo de Regresión Lineal
    X = df[["HorasEstudio"]].values  # variable independiente
    y = df["Calificacion"].values    # variable dependiente

    modelo = LinearRegression()
    modelo.fit(X, y)

    pendiente = modelo.coef_[0]
    intercepto = modelo.intercept_
    r2 = modelo.score(X, y)   # Coeficiente de determinación R^2

    # 5) Coeficiente de correlación de Pearson
    r, p_value = pearsonr(df["HorasEstudio"], df["Calificacion"])

    # 6) Gráfico de dispersión + línea de regresión
    plt.figure()
    plt.scatter(df["HorasEstudio"], df["Calificacion"], label="Datos")

    # Predicciones del modelo para la línea de regresión
    x_linea = np.linspace(df["HorasEstudio"].min(), df["HorasEstudio"].max(), 100)
    y_linea = modelo.predict(x_linea.reshape(-1, 1))

    plt.plot(x_linea, y_linea, label="Línea de regresión")
    plt.title(f"Regresión Lineal - {nombre_grupo}")
    plt.xlabel("Horas de Estudio")
    plt.ylabel("Calificación")
    plt.legend()
    plt.tight_layout()
    plt.show()

    # 7) Impresión de resultados numéricos
    print("\n--- Resultados del modelo de regresión ---")
    print(f"Pendiente (coeficiente de regresión): {pendiente:.4f}")
    print(f"Intercepto: {intercepto:.4f}")
    print(f"Ecuación de la recta: Calificación = {pendiente:.4f} * HorasEstudio + {intercepto:.4f}")

    print("\n--- Correlación de Pearson ---")
    print(f"r (coeficiente de correlación): {r:.4f}")
    print(f"p-value: {p_value:.4e}")
    print(f"R^2 del modelo: {r2:.4f}")

    # 8) Interpretaciones automáticas
    fuerza = interpretar_fuerza_correlacion(r)
    direccion = "positiva" if r > 0 else "negativa" if r < 0 else "nula"
    significativo = p_value < 0.05

    print("\n--- Interpretación automática ---")

    # 1. Interpretación del coeficiente de regresión
    print("\n1) Interpretación del coeficiente de regresión:")
    if pendiente > 0:
        print(f"   Por cada hora adicional de estudio, la calificación esperada aumenta en aproximadamente {pendiente:.2f} puntos.")
    elif pendiente < 0:
        print(f"   Por cada hora adicional de estudio, la calificación esperada DISMINUYE en aproximadamente {pendiente:.2f} puntos (relación inversa).")
    else:
        print("   La pendiente es 0, por lo que el tiempo de estudio no afecta la calificación según este modelo.")

    # 2. Interpretación del coeficiente de correlación
    print("\n2) Interpretación del coeficiente de correlación:")
    print(f"   La correlación es {direccion} y de fuerza {fuerza} (r = {r:.4f}).")

    # 3. ¿El modelo de regresión lineal es estadísticamente significativo?
    print("\n3) ¿El modelo de regresión lineal es estadísticamente significativo?")
    if significativo:
        print(f"   Sí. El p-value = {p_value:.4e} < 0.05, por lo que la relación es estadísticamente significativa.")
    else:
        print(f"   No. El p-value = {p_value:.4e} ≥ 0.05, por lo que no se puede afirmar que exista una relación lineal significativa.")

    # 4. Confiabilidad del modelo para predecir calificaciones
    print("\n4) Confiabilidad del modelo para predecir calificaciones:")
    if significativo and abs(r) >= 0.8:
        print("   El modelo es muy confiable para predecir las calificaciones basándose en las horas de estudio (alta correlación y significancia).")
    elif significativo and abs(r) >= 0.5:
        print("   El modelo es moderadamente confiable: existe relación significativa, pero no es extremadamente fuerte.")
    elif significativo:
        print("   Aunque el modelo es estadísticamente significativo, la fuerza de la relación es baja, por lo que la predicción es limitada.")
    else:
        print("   El modelo no es estadísticamente significativo, por lo que NO es confiable para predecir calificaciones.")

    print("\nConclusiones generales para este grupo:")
    print("   - Revisa el diagrama de dispersión y la línea de regresión para ver visualmente la relación.")
    print("   - Usa el valor de r, el p-value y R^2 para argumentar tus conclusiones en el reporte.")
    print("\n")


# -------------------------------------------------------------------
# PROGRAMA PRINCIPAL: analiza los tres grupos
# -------------------------------------------------------------------
if __name__ == "__main__":
    grupos = {
        "Grupo 1": "grupo1.csv",
        "Grupo 2": "grupo2.csv",
        "Grupo 3": "grupo3.csv"
    }

    for nombre, archivo in grupos.items():
        analizar_grupo(nombre, archivo)
```
