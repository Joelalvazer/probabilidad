```python

# **Probabilidad y estadística**

## Unidad 1

### Practica 5: Estadística descriptiva Dataset Autos.
### Facilitador: *Dr. José Gabriel Rodríguez Rivas*

### Alumno: Joel Avalos Alvarez
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy import stats


file_path = "/content/sample_data/autos_corregidos.csv"
autos = pd.read_csv(file_path)


precios = autos["price"].dropna()


valor_min = precios.min()
valor_max = precios.max()
moda = stats.mode(precios, keepdims=True).mode[0]
media = precios.mean()
mediana = precios.median()
desv_std = precios.std()
describe_stats = precios.describe()

print("=== Estadísticas de Precios ===")
print(f"Valor mínimo: {valor_min}")
print(f"Valor máximo: {valor_max}")
print(f"Moda: {moda}")
print(f"Media: {media}")
print(f"Mediana: {mediana}")
print(f"Desviación estándar: {desv_std}")
print("\nEstadísticos con describe():\n", describe_stats)


n = len(precios)
bins_sturges = 1 + int(math.log2(n))

plt.hist(precios, bins=bins_sturges, edgecolor="black")
plt.title("Histograma de Precios de Automóviles")
plt.xlabel("Precio (USD)")
plt.ylabel("Frecuencia")
plt.show()

print(f"Cantidad de bins según regla de Sturges: {bins_sturges}")


plt.figure(figsize=(8,5))
autos.boxplot(column="price", by="body-style", grid=False)
plt.title("Relación entre Precio y Tipo de Vehículo (Body Style)")
plt.suptitle("")
plt.xlabel("Tipo de Vehículo (Body Style)")
plt.ylabel("Precio (USD)")
plt.show()


plt.figure(figsize=(8,5))
autos.boxplot(column="price", by="drive-wheels", grid=False)
plt.title("Relación entre Precio y Tracción (Drive Wheels)")
plt.suptitle("")
plt.xlabel("Tracción (Drive Wheels)")
plt.ylabel("Precio (USD)")
plt.show()


plt.figure(figsize=(8,5))
autos.boxplot(column="price", by="engine-location", grid=False)
plt.title("Relación entre Precio y Ubicación del Motor (Engine Location)")
plt.suptitle("")
plt.xlabel("Ubicación del Motor")
plt.ylabel("Precio (USD)")
plt.show()


print("\n=== Conclusiones Generales ===")
print("1. Los precios de los autos varían entre valores accesibles (~5,000 USD) hasta autos de lujo (~45,000 USD).")
print("2. La mayoría de los autos tienen precios concentrados entre ~7,000 y ~16,500 USD (rango intercuartílico).")
print("3. El histograma muestra que los precios no están distribuidos de manera uniforme, sino que hay concentración en autos económicos.")
print("4. El boxplot por tipo de vehículo muestra que los convertibles y sedanes tienden a tener precios más altos en promedio.")
print("5. El boxplot por tracción revela que los autos con tracción trasera (rwd) presentan precios más altos en general que los de tracción delantera (fwd).")
print("6. La ubicación del motor muestra una fuerte diferencia: los autos con motor en la parte trasera tienen precios significativamente más altos que los de motor frontal.")
print("En general, los factores técnicos como tipo de vehículo, tracción y ubicación del motor influyen directamente en el precio final del\xa0automóvil.")
```
