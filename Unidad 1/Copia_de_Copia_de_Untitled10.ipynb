```python
import random
import matplotlib.pyplot as plt
from statistics import *

fabrica1 = []
for i in range (100):
  numero_aleatorio = random.uniform(50, 100)
  fabrica1.append(numero_aleatorio)

fabrica2 = []
for i in range (100):
  numero_aleatorio = random.uniform(40, 93)
  fabrica2.append(numero_aleatorio)

plt.boxplot(fabrica1)
plt.ylabel('produccion en mts.')
plt.show()

plt.boxplot(fabrica2)
plt.ylabel('produccion en mts.')
plt.show()

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))
bplot1 = ax1.boxplot(fabrica1,
                     vert=True, #aliniacion vertical
                     patch_artist=True) #rellenar color
ax1.set_title('produccion fabrica1')

bplot2 = ax2.boxplot(fabrica2,
                     vert=True, #aliniacion vertical
                     patch_artist=True) #rellebar color
ax2.set_title('produccion fabrica2')
plt.show()

fabrica1.append(150)
print(fabrica1)

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))
plot1 = ax1.boxplot(fabrica1,
                    vert=True,
                    patch_artist=True)
ax1.set_title('produccion fabrica1')
plt.show()
```
