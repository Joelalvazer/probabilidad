```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
X = np.array([0, 1, 2, 3, 4, 5])
P = np.array([0.10, 0.25, 0.30, 0.20, 0.10, 0.05])

# 1. Verificar suma de probabilidades
print("Suma de probabilidades:", np.sum(P))

# 2. Valor esperado
E_X = np.sum(X * P)
print("Valor esperado E(X):", E_X)

# 3. Varianza y desviación estándar
Var_X = np.sum((X - E_X)**2 * P)
Desv_X = np.sqrt(Var_X)
print("Varianza:", Var_X)
print("Desviación estándar:", Desv_X)

# 4. Función de distribución acumulada
F = np.cumsum(P)
print("Función acumulada F(x):", F)

# 5. Gráfica
plt.step(X, F, where='post', label='F(x)')
plt.xlabel('Número de células anormales')
plt.ylabel('Probabilidad acumulada')
plt.title('Función de distribución acumulada - Ejercicio 1')
plt.grid(True)
plt.legend()
plt.show()
```
