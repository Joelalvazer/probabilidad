```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
X3 = np.array([0, 1, 2, 3, 4, 5])
P3 = np.array([0.05, 0.10, 0.25, 0.30, 0.20, 0.10])

# 1. Verificar suma
print("Suma de probabilidades:", np.sum(P3))

# 2. Valor esperado
E_X3 = np.sum(X3 * P3)
print("Valor esperado E(X):", E_X3)

# 3. Varianza y desviación estándar
Var_X3 = np.sum((X3 - E_X3)**2 * P3)
Desv_X3 = np.sqrt(Var_X3)
print("Varianza:", Var_X3)
print("Desviación estándar:", Desv_X3)

# 4. Función acumulada
F3 = np.cumsum(P3)
print("Función acumulada F(x):", F3)

# 5. Gráfica
plt.step(X3, F3, where='post', label='F(x)')
plt.xlabel('Unidades vendidas')
plt.ylabel('Probabilidad acumulada')
plt.title('Función de distribución acumulada - Ejercicio 3')
plt.grid(True)
plt.legend()
plt.show()
```
