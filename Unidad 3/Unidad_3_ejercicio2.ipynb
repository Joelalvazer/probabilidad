```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
X2 = np.array([0, 10, 20, 50])
P2 = np.array([0.50, 0.30, 0.15, 0.05])

# 1. Verificar suma
print("Suma de probabilidades:", np.sum(P2))

# 2. Valor esperado
E_X2 = np.sum(X2 * P2)
print("Valor esperado E(X):", E_X2)

# 3. Varianza y desviación estándar
Var_X2 = np.sum((X2 - E_X2)**2 * P2)
Desv_X2 = np.sqrt(Var_X2)
print("Varianza:", Var_X2)
print("Desviación estándar:", Desv_X2)

# 4. Función acumulada
F2 = np.cumsum(P2)
print("Función acumulada F(x):", F2)

# 5. Gráfica
plt.step(X2, F2, where='post', label='F(x)')
plt.xlabel('Monto del premio (pesos)')
plt.ylabel('Probabilidad acumulada')
plt.title('Función de distribución acumulada - Ejercicio 2')
plt.grid(True)
plt.legend()
plt.show()
```
